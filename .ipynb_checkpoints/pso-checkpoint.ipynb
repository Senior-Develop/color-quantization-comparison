## Define commonly used functions here

```python
#recreate image with particles in swarm
def recreate_image(codebook, labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            #get color of pixel(i,j) with the particle and label created from that particle
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image
#define a function calculating mean square error between two different images
def mse(imageA, imageB):
    # the 'Mean Squared Error' between the two images is the
    # sum of the squared difference between the two images;
    # NOTE: the two images must have the same dimension
    err = np.sum((imageA.astype("float") * 255 - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])

    # return the MSE, the lower the error, the more "similar"
    # the two images are
    return err
def mse1(imageA, imageB):
    # the 'Mean Squared Error' between the two images is the
    # sum of the squared difference between the two images;
    # NOTE: the two images must have the same dimensionk
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])

    # return the MSE, the lower the error, the more "similar"
    # the two images are
    return err
def get_unique_colors(img):
    return np.unique(img.reshape(-1, img.shape[-1]), axis=0, return_counts=True)
```

# Color Quantization using median-cut

```python
class MCutQuantization:
    def __init__(self, n_colors, image):  
        #initalize kmeans parameters with user defined parameters
        self.image = image
        self.n_colors = n_colors
        self.depth = int(math.log(n_colors, 2))
        w, h, d = original_shape = tuple(image.shape)
        self.result = np.zeros((w, h, d))
        
    def flat_image(self, image):
        flattened_img_array = []
        for rindex, rows in enumerate(image):
            for cindex, color in enumerate(rows):
                flattened_img_array.append([color[0],color[1],color[2],rindex, cindex]) 

        flattened_img_array = np.array(flattened_img_array)
        self.split_into_buckets(image, flattened_img_array, self.depth)        
    def median_cut_quantize(self, img, img_arr):
        # when it reaches the end, color quantize
        r_average = np.mean(img_arr[:,0])
        g_average = np.mean(img_arr[:,1])
        b_average = np.mean(img_arr[:,2])
        for data in img_arr:
            self.result[data[3]][data[4]] = [r_average, g_average, b_average]
    
    def split_into_buckets(self, img, img_arr, depth):
        if len(img_arr) == 0:
            return 
        if depth == 0:
            self.median_cut_quantize(img, img_arr)
            return
        r_range = np.max(img_arr[:,0]) - np.min(img_arr[:,0])
        g_range = np.max(img_arr[:,1]) - np.min(img_arr[:,1])
        b_range = np.max(img_arr[:,2]) - np.min(img_arr[:,2])
        
        space_with_highest_range = 0

        if g_range >= r_range and g_range >= b_range:
            space_with_highest_range = 1
        elif b_range >= r_range and b_range >= g_range:
            space_with_highest_range = 2
        elif r_range >= b_range and r_range >= g_range:
            space_with_highest_range = 0

        # sort the image pixels by color space with highest range 
        # and find the median and divide the array.
        img_arr = img_arr[img_arr[:,space_with_highest_range].argsort()]
        median_index = int((len(img_arr)+1)/2)

        #split the array into two blocks
        self.split_into_buckets(img, img_arr[0:median_index], depth-1)
        self.split_into_buckets(img, img_arr[median_index:], depth-1)
        
    def optimize(self):        
        t0 = time()
        self.flat_image(self.image)
        unique_color_count = len(get_unique_colors(self.result)[0])
        print("Kmeans - recreated image with unique colors {}".format(unique_color_count))
        self.result = self.result.astype(np.uint8)
        mse_err = mse1(self.result, self.image)        
        io.imshow(self.result)
        io.imsave(fname="output/mcut/mcut.jpeg", arr=self.result)
        print("Kmeans - computational time {}".format(time() - t0))
        print("Kmeans - MSE error {}".format(mse_err))
        
```

# Color Quantization using K-Means

Performs a pixel-wise Vector Quantization (VQ) of an image of an image
, reducing the number of colors required to show the image from 96,615
unique colors to 64, while preserving the overall appearance quality.

In this example, pixels are represented in a 3D-space and K-means is used to
find user-defined quantization color clusters. In the image processing literature,
the codebook obtained from K-means (the cluster centers) is called the color palette.
Using a single byte, up to 256 colors can be addressed, whereas an RGB encoding
requires 3 bytes per pixel. The GIF file format, for example, uses such a
palette.

For comparison, a quantized image using a random codebook (colors picked up
randomly) is also shown.

```python
class KmeansQuantization:
    def __init__(self, n_colors, image):  
        #initalize kmeans parameters with user defined parameters
        self.image = image
        self.n_colors = n_colors
    def optimize(self):
        t0 = time()
        # Convert to floats instead of the default 8 bits integer coding. Dividing by
        # 255 is important so that plt.imshow behaves works well on float data (need to
        # be in the range [0-1])
        img = np.array(self.image, dtype=np.float64) / 255

        # Load Image and transform to a 2D numpy array.
        w, h, d = original_shape = tuple(img.shape)
        assert d == 3
        image_array = np.reshape(img, (w * h, d))
        image_array_sample = shuffle(image_array, random_state=0)[:1000]
        kmeans = KMeans(n_clusters=self.n_colors, random_state=0).fit(image_array_sample)

        # Get labels for all points
        labels = kmeans.predict(image_array)            
        codebook_random = shuffle(image_array, random_state=0)[:self.n_colors]
        labels_random = pairwise_distances_argmin(codebook_random, image_array, axis=0)
        
        recreated_image = recreate_image(kmeans.cluster_centers_, labels, w, h)
        mse_err = mse(recreated_image, self.image)
        recreated_image = (recreated_image * 255).astype(np.uint8)
        unique_color_count = len(get_unique_colors(recreated_image)[0])        
        print("Kmeans - recreated image with unique colors {}".format(unique_color_count))
        io.imshow(recreated_image)
        print("Kmeans - computational time {}".format(time() - t0))
        print("Kmeans - MSE error {}".format(mse_err))
        io.imsave(fname="output/kmeans/kmeans.jpeg", arr=recreated_image)
#         return recreated_image
```

# How we can use PSO in image quantization

## Quantization overview
Let us say we are going to quantize an image with 64 colors
Obiously, what we need most is a color pallette with 64 different colors and this color pallette means 64 X 3 type matrix mathematically.
So at this moment we can define the image quantization as defining most approximate color pallette and matching each pixel in an image with the color in the pallette.
- pallette array
[0, 1, 4], [1, 9 , 6], ... (length of pallette = 64:quantization image)
- label array
[23, 0, 7, 24, 1, ...] (length of label = width * height)
So the different quantization type means different ways of searching pallettes.

## Then how we can decide pallette in PSO quantization?

First, In PSO color quantization we say a candidate pallette as a particle.
    So in our example a particle is a 2d array of 64 * 3.
    like this:
         [
             [0, 1, 4],
             [1, 9, 6],
             ...
             [254, 255, 251]
         ]
         (totally 64 3d vectors involved in above array)
         
Secondly, we define a swarm of these color pallette particles.
    i.e let's say we defined a swarm with 10 particles then the shape of that swarm will be 10 * 64 * 3
    We will make use of this swarm to redraw an image.
    What redrawing means exactly is labeling eaching pixel with the most close color in a color pallette particle.
    This will iterate every particle in the swarm.
    
Thirdly, We will update the swarm it self.
    Then how do we update the swarm?
    We update the swarm by updating each particles in the swarm and update global best value of the swarm.
    Then how do we update a particle?
    We update the particle according to the [formula](https://pyswarms.readthedocs.io/en/latest/api/pyswarms.single.html#module-pyswarms.single.global_best) of pso.  
    Here our purpose is to make the quantized image as close as possible so our objective function will be calculating mse value of original image and quantized image.
    We will update the position and velocity of the particle as mse value decreases.
    
Finally, when mse value of particle's current position is smaller than its private best value we update its private value as its current position.
    And if a particle's private best value's mse value is smaller than the swarm's global gbest value we will update the swarm's global best as a particle's private best value.
    All these processes will be repeated during the iterations a user defined.
        
        

```python
# class to define a swarm of color pallette particles
# n_color=quantization color, image=original image, swarm_size=swarm's particl number, pkmeans=percent of random images in an image
class ColorSwarm:
    def __init__(self, n_colors, image, swarm_size, pkmeans=0.1):  
        #initalize swarm parameters with user defined parameters
        self.swarm = {}
        self.n_colors = n_colors
        self.image = image
        self.pkmeans = pkmeans
        self.swarm_size= swarm_size
        print("swarm_size is {}".format(swarm_size))
        self.image_array_normalized = np.array(image, dtype=np.float64) / 255
        self.w, self.h, self.d = original_shape = tuple(self.image_array_normalized.shape)
        self.image_array_reshaped = np.reshape(self.image_array_normalized, (self.w * self.h, self.d))
        self.options = {'c1':1.49, 'c2':1.49, 'w':0.72, 'r1':1, 'r2':1}
        #initialize swarm
        self.init_swarm(swarm_size)
        self.count = 1
    #initialize swarm parameters
    def init_options(self, options):
        self.options = options
        print("changed options", self.options)
    #initialize swarm
    def init_swarm(self, swarm_size):
        #initialize particles
        particles = self.init_particles(swarm_size)
        #initialize swarm's particle option with particles above
        self.swarm['particles'] = particles
        print("length of self.swarm['particles'] is {}".format(len(self.swarm['particles'])))
        #initialize swarm's global best value with random 64*3 array from the original image
        start = int(self.image_array_reshaped.shape[0] * np.random.rand())
        self.swarm['gbest'] = self.image_array_reshaped[start:start + self.n_colors].copy()
    #init paricles
    def init_particles(self, swarm_size, pmin=0, pmax=1, vmax=50):
        particles = []
        for i in range(swarm_size):
            #init particle's current position from kmeans centroid values to reduce search space
            image_array_sample = shuffle(self.image_array_reshaped, random_state=i)[:1000]
#             print("image_array_sample ", image_array_sample)
            kmeans = KMeans(n_clusters=n_colors, random_state=0).fit(image_array_sample)
#             print((kmeans.cluster_centers_).shape)
            particle = {}
            particle['current_position'] = kmeans.cluster_centers_.copy()
            #init paritcles's velocity using random 64*3 array
            particle['velocity'] = np.random.rand(self.n_colors, self.d) * vmax / 255
            #init a paritcle's private best value with its current position which was derived from kmeans centroid
            particle['pbest'] = particle['current_position'].copy()
            #save each particle to the particle list
            particles.append(particle)        
        return particles
    #define a function to label each pixels with quantized color pallete which means particle's current position
    def f(self, position):
        labels_random = pairwise_distances_argmin(position, self.image_array_reshaped, axis=0)
        recreated_image = recreate_image(position, labels_random, self.w, self.h)        
        mse_err = mse(recreated_image, self.image)
        return mse_err
    #define a function to draw an image in pso directory for each iteration with swarm's global best value
    def draw(self, position):
        labels_random = pairwise_distances_argmin(position, self.image_array_reshaped, axis=0)
        print("unique color count in quantized image is {}".format(len(np.unique(labels_random))))
        recreated_image = recreate_image(position, labels_random, self.w, self.h)
        temp = (recreated_image * 255).astype(np.uint8)
        self.count += 1
        Image.fromarray(temp, 'RGB').save("output/pso/{}.jpeg".format(self.count), "jpeg")      
        mse_err = mse(recreated_image, self.image)
        unique_color_array = np.unique(temp.reshape(-1, temp.shape[-1]), axis=0, return_counts=True)
        return mse_err
    #update position and velocity of a particle using PSO's update formula
    def update_state(self):
        for i in range(self.swarm_size):
            self.swarm['particles'][i]['velocity'] = (self.options['w'] * self.swarm['particles'][i]['velocity']
            + self.options['c1'] * (self.swarm['particles'][i]['pbest'] - self.swarm['particles'][i]['current_position'])
            + self.options['c2'] * (self.swarm['gbest'] - self.swarm['particles'][i]['pbest'])).copy()
            
            self.swarm['particles'][i]['current_position'] += self.swarm['particles'][i]['velocity']
            self.swarm['particles'][i]['current_position'][self.swarm['particles'][i]['current_position'] < 0] = 0
            self.swarm['particles'][i]['current_position'][self.swarm['particles'][i]['current_position'] > 1] = 1
            self.swarm['particles'][i]['current_position'] = self.swarm['particles'][i]['current_position'].copy()
    #optimze function with iterations defined
    def optimize(self, iterations=100):
        #record start time for optimzation
        start = time()
        print("length of swarm is {}".format(len(self.swarm['particles'])))
        #iterating during the iterations
        for i in range(iterations):
            #call draw function for drawing quantized picture
            mse = self.draw(self.swarm['gbest'])        
            print("gbest mse ------------------------------->>>>>>>>>>>>>>>>>>>>>>>> ", mse )
            print("psnr is {}".format(20 * math.log10(255 / math.sqrt(mse))))
            #call update function for update particle's position and veloctiy
            self.update_state()            
            #iterate over the swarm
            for i in range(self.swarm_size):
                
                current_position_mse = self.f(self.swarm['particles'][i]['current_position'])
                pbest_mse = self.f(self.swarm['particles'][i]['pbest'])
                gbest_mse = self.f(self.swarm['gbest'])
                #check if particle's current position is smaller than its private best value
                if current_position_mse < pbest_mse:
                    #update pbest
                    self.swarm['particles'][i]['pbest'] = self.swarm['particles'][i]['current_position'].copy()     
                #check if a particle's pbest is smaller than swarm's gbest
                if pbest_mse < gbest_mse:
                    #update gbest
                    self.swarm['gbest'] = self.swarm['particles'][i]['pbest'].copy()
                    print("------------------------------------------pallette is now updated!------------------------------------------")
                print("mse_err is {} {}".format(current_position_mse, pbest_mse))
        #record end time and calcuate overall computational time
        end = time()
        print("PSO - computational time {}".format(end - start))
        return self.swarm['gbest']
        
                
```
